```python
%matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import dataset

filepath_MCF_Unf = "raw_data\MCF7_SmartS_Unfiltered_Data.txt"
df_MCF_Unf = pd.read_csv(filepath_MCF_Unf,delimiter="\ ",engine='python',index_col=0)
print("Dataframe dimensions:", np.shape(df_MCF_Unf))
```

```python
df_MCF_Unf.iloc[:3,:]
```

```python
df_MCF_Unf.index[2]
```

```python
#List of all cells that were part of the hypoxia and normoxia groups
hypo_MCF = []
norm_MCF = []
for cell in df_MCF_Unf.columns:
    if "Hypo" in cell.split("_"):
        hypo_MCF.append(cell)
    elif "Norm" in cell.split("_"):
        norm_MCF.append(cell)
    else:
        print("Unkown:", cell)

print("Hypo:", len(hypo_MCF), "Norm:", len(hypo_MCF))

#Data sets that contain only hypoxia cells
df_Hypo = df_MCF_Unf[hypo_MCF]
#Data sets that contain only hypoxia cells
df_Norm = df_MCF_Unf[norm_MCF]
```

```python
df_Norm.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x1 = df_Norm.loc['"ACTG1"']
y1 = df_Norm.loc['"GAPDH"']
z1 = df_Norm.loc['"ALDOA"']

x2 = df_Hypo.loc['"ACTG1"']
y2 = df_Hypo.loc['"GAPDH"']
z2 = df_Hypo.loc['"ALDOA"']

ax.set_xlabel(df_MCF_Unf.index[0])
ax.set_ylabel(df_MCF_Unf.index[1])
ax.set_zlabel(df_MCF_Unf.index[2])

ax.scatter(x1, y1, z1, c='g', label='Normoxia')
ax.scatter(x2, y2, z2, c='r', label='Hypoxia')

plt.legend()
plt.show()
```
